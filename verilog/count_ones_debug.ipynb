```python
def count_ones_in_chunks(file_path):
    chunk_size = 800
    num_chunks = 10

    with open(file_path, 'r') as f:
        for line_num, line in enumerate(f, start=1):
            line = line.strip()
            if len(line) != chunk_size * num_chunks:
                print(f"Warning: Line {line_num} is not {chunk_size * num_chunks} bits long.")
                continue

            counts = [line[i:i+chunk_size].count('1') for i in range(0, len(line), chunk_size)]
            print(f"Line {line_num}: {counts}")

# Example usage
count_ones_in_chunks("C:/projects/npc25-difflogic-4-control/verilog/debug.txt")
```
